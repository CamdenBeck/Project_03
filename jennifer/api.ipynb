```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time
from scipy.stats import linregress
import requests
import json
# Import the API key
from config import flight_key
from pprint import pprint
```

```python

# Specify the URL
schedule_url = "https://api.flightapi.io/schedule"

# Define parameters for the request
iata_code = "LAX"  
day = 1  # Example: First day of the month

# Construct the full API URL
api_url = f"{schedule_url}/{flight_key}?mode=departures&iata={iata_code}&day={day}"

# Make the GET request
response = requests.get(api_url)

#from chatgpt
# Check if the request was successful
if response.status_code == 200:
    # Pretty-print the JSON response
    pprint(response.json())
else:
    print(f"Error: {response.status_code} - {response.text}")
```

```python
trackbyroute_url = "https://api.flightapi.io/trackbyroute"

# Define the request parameters
date = "20240128"  # Example date (YYYYMMDD)
airport1 = "lax"   # Los Angeles International Airport
airport2 = "jfk"   # John F. Kennedy International Airport (you can change this)

# Construct the full API URL with parameters
api_url = f"{trackbyroute_url}/{flight_key}?date={date}&airport1={airport1}&airport2={airport2}"

# Make the GET request
response = requests.get(api_url)

# Check if the request was successful
if response.status_code == 200:
    # Pretty-print the JSON response
    pprint(response.json())
else:
    print(f"Error: {response.status_code} - {response.text}")
```
